```python
from sklearn.svm import SVR
import pandas_datareader as web
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
end=datetime.date.today()
df=web.DataReader('KIRLFER.NS',data_source='yahoo', start = '2019-01-02', end = end)
aa=27
df = df.tail(aa)
df
```

```python
plt.figure(figsize=(16,8))
plt.title("Close Price History")
plt.plot(df['Close'])
plt.xlabel('Date',fontsize=18)
plt.ylabel('Close Price USD',fontsize=18)
plt.grid(True)
plt.show()
```

```python
close = df.filter(['Close'])
close
```

```python
n = []
for i in range(aa):
    n.append(i+1)
n
```

```python
close['Numbers'] = n
```

```python
close
```

```python
X=close.iloc[:,1:2].values
y=close.iloc[:,:1].values
```

```python
X
```

```python
y
```

```python
y.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
sc_y=StandardScaler()
X=sc_x.fit_transform(X)
y=sc_y.fit_transform(y)
```

```python
X
```

```python
y
```

```python
from sklearn.svm import SVR
```

```python
lin_svr = SVR(kernel = 'linear', C=1000.0)
lin_svr.fit(X,y)

poly_svr = SVR(kernel = 'poly', C=1000.0, degree = 2)
poly_svr.fit(X,y)

rbf_svr = SVR(kernel = 'rbf', C=1000.0, gamma = 0.15)
rbf_svr.fit(X,y)
```

```python
#plt.figure(figsize=(16,8))
plt.scatter(sc_x.inverse_transform(X),sc_y.inverse_transform(y),color = 'red', label = 'Data')
plt.plot(sc_x.inverse_transform(X),sc_y.inverse_transform(rbf_svr.predict(X)),color = 'green', label = 'RBF Model')
plt.plot(sc_x.inverse_transform(X),sc_y.inverse_transform(poly_svr.predict(X)),color = 'orange', label = 'POLY Model')
plt.plot(sc_x.inverse_transform(X),sc_y.inverse_transform(lin_svr.predict(X)),color = 'blue', label = 'LIN Model')
plt.legend()
plt.show()
```

```python
R=sc_y.inverse_transform(rbf_svr.predict(sc_x.transform([[aa+1]])))
P=sc_y.inverse_transform(poly_svr.predict(sc_x.transform([[aa+1]])))
L=sc_y.inverse_transform(lin_svr.predict(sc_x.transform([[aa+1]])))
print(R, P, L)
```
